```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
  
# a = np.array([1, 2, 3])
# b = np.array([4, 5, 6, 7])

x = np.arange(1024)
y = np.arange(1024)

x, y = np.meshgrid(x, y)

  
fig = plt.figure(figsize=(10,10))
axes = fig.gca(projection ='3d')
axes.plot_surface(x, y, c)
plt.savefig('/scratch/gilbreth/akamsali/Research/Makin/Auditory_Cortex/notebooks/down/surface_plt.svg')

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(111, projection='3d')
ax1.set_title('Unnormalised')
surf1 = ax1.plot_surface(x, y, c/np.max(c, axis=0), cmap=cm.ocean)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Un-Normalised value')
# ax1.set_zlabel('Normalised value')
fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=5)
plt.savefig('/scratch/gilbreth/akamsali/Research/Makin/Auditory_Cortex/notebooks/down/surface_plt_ocean.svg')

  
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(111, projection='3d')
ax1.set_title('Normalised')
surf1 = ax1.plot_surface(x, y, c/np.max(c), cmap=cm.ocean)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Normalised value')
fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=5)
plt.savefig('/scratch/gilbreth/akamsali/Research/Makin/Auditory_Cortex/notebooks/down/surface_plt_ocean_norm.svg')

plt.show()
```
