#### Summary:
- deepspeech2: spectrogram normalized to have zero mean and unit variance (globally)
- speech2text: spectrogram normalized to have zero mean and unit variance (separately along each frequency)
- whisper (tiny & base): In the paper they say, they globally scale the input to lie between -1 and 1 and approximately having zero-mean. But I tested it for random sentences, I could not see this normalization...
- wav2vec2: normalizes input audio to zero-mean and unit variance
