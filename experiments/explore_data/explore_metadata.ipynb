```python
def get_avg_corr(long_f, raw_f):
    avg_corr = 0
    num_units = long_f.shape[1]
    for n in range(num_units):
        avg_corr += np.corrcoef(long_f[:,n], raw_f[:,n])[0,1]
    avg_corr /= num_units
    return avg_corr

def get_corr_vs_time(long_f, raw_f):
    corrs = []
    time_steps = long_f.shape[0]
    for i in range(time_steps):
        corrs.append(np.corrcoef(long_f[i,:], raw_f[i,:])[0,1])
    return np.array(corrs)

def get_layer_correlations_for_stim_number(model_name, raw_features, stim_number=12):
    num_previous_stimuli=9

    long_features, last_sent_ID, total_duration = dataloader.get_features_for_long_stimulus(
        model_name=model_name, stim_number=stim_number, num_previous_stimuli=num_previous_stimuli
    )

    # sent_duration = dataloader.metadata.stim_duration(sent_ID)
    bin_width = 20/1000.0
    n = int(np.ceil(round(total_duration/bin_width, 3)))
    # print(total_duration)
    # print(n)


    layer_wise_corr = {}
    print(f"For model: {model_name}, sent_ID: {last_sent_ID}")
    for layer in raw_features.keys():
        num_features, num_units = raw_features[layer][last_sent_ID].shape
        raw_features_last_sent = raw_features[layer][last_sent_ID]
        # print(num_features)
        long_features_resampled = signal.resample(long_features[layer], n, axis=0)


        long_features_last_sent = long_features_resampled[-num_features:]
        # print(long_features_last_sent.shape)
        # avg_corr = get_avg_corr(long_features_last_sent, raw_features_last_sent)
        # print(f"Correlation among fearures for layer: {layer} = {avg_corr:.3f}")
        
        layer_wise_corr[layer] = get_corr_vs_time(long_features_last_sent, raw_features_last_sent)
    
        plt.plot(layer_wise_corr[layer], label=f'L-{layer}')
    plt.legend(loc='best')
    plt.title(f"{model_name}, sent-{last_sent_ID}")
        
```
